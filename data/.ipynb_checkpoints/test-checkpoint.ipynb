```python
features = []

cols = ['neg', 'neu', 'pos', 'compound']

for row, item in train_csv.iterrows():
    scores = sid.polarity_scores(item["transcript"])
    features.append([scores[col] for col in cols])

train_csv[cols] = features
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

def add_nltk_features(df):
    features = []
    cols = ['neg', 'neu', 'pos', 'compound']

    for row, item in df.iterrows():
        scores = sid.polarity_scores(item["transcript"])
        features.append([scores[col] for col in cols])

    df[cols] = features
```

```python
add_nltk_features(train_csv)
train_csv
```

```python
train_csv[['neg', 'neu', 'pos', 'compound']] = train_csv.apply(
    lambda item: [sid.polarity_scores(item['transcript'])[col] for col in cols], axis='columns').to_list()
```
